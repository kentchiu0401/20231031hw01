```python
import requests
from bs4 import BeautifulSoup

url = 'https://tw.sports.yahoo.com/nba/'

try:
    response = requests.get(url, timeout=5)
    response.raise_for_status()  # 檢查網頁請求是否成功
    soup = BeautifulSoup(response.content, 'html.parser')

    datas = []

    # 抓取新聞標題、摘要、超連結和圖片連結
    news_articles = soup.find_all('li', class_='js-stream-content')
    for article in news_articles:
        title = article.find('h3').text.strip()
        summary = article.find('p').text.strip()
        link = article.find('a')['href']
        image_link = article.find('img')['src']

        # 將資訊儲存到datas列表中
        datas.append({
            '標題': title,
            '摘要': summary,
            '新聞連結': link,
            '圖片連結': image_link
        })

    # 輸出結果，每條新聞資訊分行顯示
    for data in datas:
        print(f"標題: {data['標題']}")
        print(f"摘要: {data['摘要']}")
        print(f"新聞連結: {data['新聞連結']}")
        print(f"圖片連結: {data['圖片連結']}")
        print('-' * 80)  # 分隔每則新聞

except requests.exceptions.RequestException as e:
    print(f"網頁爬取失敗: {e}")

except Exception as e:
    print(f"發生未知的錯誤: {e}")
    
import pandas as pd
df = pd.DataFrame(datas,columns=['標題','摘要','新聞連結','圖片連結'])
df
```

```python
df.to_csv('Yahoo運動NBA.csv',encoding='utf-8-sig')
```
